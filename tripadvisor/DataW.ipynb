```python
import numpy as np
import pandas as pd
import re
```

```python
trip = pd.read_csv('./comments_pag.csv')
trip
```

```python
trip.info()
```

## Name

```python
trip.name = trip.name.apply(lambda x: x.split("$")[0].split("-")[0])
trip
```

## Price

```python
trip.price = trip.price.apply(lambda x: int(x.replace("MX$","").replace(",","")) if str(x)!="nan" else np.nan)
trip
```

## Score

```python
trip.score = trip.score.apply(lambda x: int(x[-2:]))
trip
```

## Stars

```python
trip.stars = trip.stars.apply(lambda x: float(x.split()[0]) if str(x)!="nan" else np.nan )
trip
```

# Hotels

```python
hotels = trip[['city','name','price','quality_price','rating','stars','url']]
hotels
```

```python
hotels = hotels.drop_duplicates()
hotels["id"] = np.arange(hotels.shape[0])
```

```python
hotels
```

# Comments

```python
comments = trip[['name','comment','score']]
comments
```

```python
comments = comments.merge(hotels[["name","id"]],on="name")
comments.drop("name",axis = 1,inplace=True)
```

```python
comments = comments.rename(columns={"id":"id_hotel"})
comments
```

### Removing emojis
Credits: https://stackoverflow.com/a/58356570

```python
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
```

```python
comments.comment = comments.comment.apply(remove_emojis)
comments
```
